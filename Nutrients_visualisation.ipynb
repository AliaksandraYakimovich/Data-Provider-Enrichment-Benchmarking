Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

Loading dataset

```python
df = pd.read_csv('nutrients_csvfile.csv')
```

```python
numeric_cols = ['Grams', 'Calories', 'Protein', 'Fat', 'Sat.Fat', 'Fiber', 'Carbs'] #changing data type for numeric columns (before object(string))

for i in numeric_cols:
  df[i]=pd.to_numeric(df[i], errors='coerce')
```

Linear graph

```python
items = np.arange(1, 101)
calories = df['Calories'].head(100).values
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(items, calories, color='royalblue', marker='o', linewidth=2, label='Calories')

ax.set_title('Calories for first 100 products', fontsize=14, fontweight='bold')
ax.set_xlabel('Number of product', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend()

plt.show()
```

Distribution histogram

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(calories, bins=30, color='seagreen', edgecolor='black', alpha=0.7)

ax.set_title('Distribution of the product calories', fontsize=14)
ax.set_xlabel('Calories')
ax.set_ylabel('Number of products')

plt.show()
```

The histogram shows that the distribution of calories is not symmetrical. The vast majority of the products in the dataset have a caloric value ranging from 50 to 300 kcal, which forms the peak of the graph. Products exceeding 400 kcal are significantly less frequent that includes a few extreme outliers reaching up to approximately 1000 kcal

Scatter plot

```python
scatter_data = df[['Fat', 'Calories']].dropna()
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(scatter_data['Fat'], scatter_data['Calories'], color='darkorange', alpha=0.4, edgecolor='white', s=60)

ax.set_title('Correlation between Fat and Calories', fontsize=14, fontweight='bold')
ax.set_xlabel('Fat (grams)', fontsize=12)
ax.set_ylabel('Calories (kcal)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.3)

plt.show()
```

The scatter plot demonstrates the overall correlation between Fat and Calories.

However, visual analysis revealed critical data anomalies in the dataset. For instance, there is a data point indicating ~230g of fat with only ~200 calories, and another with ~115g of fat and ~100 calories. Physiologically, 1 gram of fat yields 9 kcal, meaning these values are physically impossible and indicate data entry errors in the source dataset.

Distributions

```python
sample_size = 1000
normal_dist = np.random.normal(loc=0, scale=1, size=sample_size)
uniform_dist = np.random.uniform(low=-3, high=3, size=sample_size)
exponential_dist = np.random.exponential(scale=1, size=sample_size)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(normal_dist, bins=30, edgecolor='black', alpha=0.7)
axes[0].set_title('Normal Distribution')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

axes[1].hist(uniform_dist, bins=30, edgecolor='black', alpha=0.7)
axes[1].set_title('Uniform Distribution')
axes[1].set_xlabel('Value')

axes[2].hist(exponential_dist, bins=30, edgecolor='black', alpha=0.7)
axes[2].set_title('Exponential Distribution')
axes[2].set_xlabel('Value')

plt.tight_layout()
plt.show()
```

Box Plot

```python
clean_calories = df['Calories'].dropna().values

fig, ax = plt.subplots(figsize=(10, 4))
ax.boxplot(clean_calories, vert=False, whis=[2, 98], patch_artist=True, medianprops=dict(color='white'))

ax.set_title('Outlier Detection', fontsize=14, fontweight='bold')
ax.set_xlabel('Calories')
ax.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
